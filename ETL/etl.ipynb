```python
import pandas as pd

# Memuat dataset dari file CSV yang diunggah
df = pd.read_csv("Superstore Sales Dataset.csv")

# Memilih kolom yang dibutuhkan
df_customer = df[['Customer ID', 'Customer Name', 'Segment', 'Country']]

# Menghapus duplikat berdasarkan Customer ID
df_customer_unique = df_customer.drop_duplicates(subset=['Customer ID'])

# Menyimpan hasilnya ke file CSV baru
df_customer_unique.to_csv("dim_customer.csv", index=False)

df_customer_unique
```

```python
import pandas as pd

# Memuat dataset dari file CSV yang diunggah
df = pd.read_csv("Superstore Sales Dataset.csv")

# Memilih kolom yang dibutuhkan
df_postal = df[['City', 'State', 'Postal Code', 'Region']]

# Menghapus duplikat berdasarkan Postal Code
df_postal_unique = df_postal.drop_duplicates(subset=['Postal Code'])

# Menyimpan hasilnya ke file CSV baru
df_postal_unique.to_csv("dim_postal.csv", index=False)

df_postal_unique
```

```python
import pandas as pd

# Memuat dataset dari file CSV yang diunggah
df = pd.read_csv("Superstore Sales Dataset.csv")

# Pilih kolom yang diperlukan untuk dimensi produk
df_product = df[['Product ID', 'Category', 'Sub-Category', 'Product Name']]

# Hapus duplikat berdasarkan Product ID
df_product_unique = df_product.drop_duplicates(subset=['Product ID'])

# Simpan hasilnya ke file CSV baru
df_product_unique.to_csv("dim_product.csv", index=False)

df_product_unique
```

```python
import pandas as pd

# Memuat dataset dari file CSV yang diunggah
df = pd.read_csv("Superstore Sales Dataset.csv")

# Pilih kolom yang diperlukan untuk tabel fakta
df_fact = df[['Order Date', 'Ship Date', 'Ship Mode', 'Customer ID', 'Postal Code', 'Product ID', 'Sales']]

# Ubah tipe data kolom Order Date dan Ship Date menjadi datetime
df_fact['Order Date'] = pd.to_datetime(df_fact['Order Date'], dayfirst=True)
df_fact['Ship Date'] = pd.to_datetime(df_fact['Ship Date'], dayfirst=True)

# Simpan hasilnya ke file CSV baru
df_fact.to_csv("fact_table.csv", index=False)
df_fact
```
